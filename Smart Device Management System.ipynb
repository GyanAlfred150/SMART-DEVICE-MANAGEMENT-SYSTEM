```python
print("===== SMART DEVICE MANAGEMENT SYSTEM =====")
class SmartDevice:
    def __init__(self, name, device_id):
        self.name = name
        self.__device_id = None
        self.__power_status = "OFF"
        self.set_device_id(device_id)

    def get_device_id(self):
        return self.__device_id

    def get_power_status(self):
        return self.__power_status

    def set_device_id(self, device_id):
        if not device_id or not device_id.strip():
            raise ValueError("Device ID cannot be empty.")
        self.__device_id = device_id.strip()

    def set_power_status(self, status):
        if status not in ("ON", "OFF"):
            raise ValueError("Power status must be 'ON' or 'OFF'.")
        self.__power_status = status

    def turn_on(self):
        self.set_power_status("ON")
        print(f"{self.name} is now ON.")

    def turn_off(self):
        self.set_power_status("OFF")
        print(f"{self.name} is now OFF.")

    def display_info(self):
        print("\n----- Device Information -----")
        print("Device Name   :", self.name)
        print("Device ID     :", self.__device_id)
        print("Power Status  :", self.__power_status)

class TemperatureSensor(SmartDevice):
    def __init__(self, name, device_id):
        super().__init__(name, device_id)
        self.temperature = 25.0      # default

    def set_temperature(self):
        try:
            self.temperature = float(input("Enter new temperature (°C): "))
            print("Temperature updated.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    def show_temperature(self):
        self.display_info()
        print("Temperature    :", self.temperature, "°C")

class SmartLight(SmartDevice):
    def __init__(self, name, device_id):
        super().__init__(name, device_id)
        self.__brightness = 50

    def get_brightness(self):
        return self.__brightness

    def set_brightness(self, value):
        try:
            value = int(value)
            if 0 <= value <= 100:
                self.__brightness = value
                print("Brightness updated.")
            else:
                print("Brightness must be between 0 and 100.")
        except ValueError:
            print("Invalid input. Please enter an integer.")

    def adjust_brightness(self):
        val = input("Enter brightness (0-100): ")
        self.set_brightness(val)

    def show_light(self):
        self.display_info()
        print("Brightness     :", self.__brightness, "%")

class SecurityCamera(SmartDevice):
    def __init__(self, name, device_id):
        super().__init__(name, device_id)
        self.__recording = False     # private

    def get_recording(self):
        return self.__recording

    def start_recording(self):
        if self.get_power_status() != "ON":
            print("Cannot start recording: Camera is OFF.")
            return
        self.__recording = True
        print("Recording started.")

    def stop_recording(self):
        self.__recording = False
        print("Recording stopped.")

    def show_camera(self):
        self.display_info()
        rec_status = "Yes" if self.__recording else "No"
        print("Recording      :", rec_status)

light = SmartLight("Living Room Light", "L001")
sensor = TemperatureSensor("Living Room Sensor", "T001")
camera = SecurityCamera("Front Door Camera", "C001")

while True:
    print("\n===== MAIN MENU =====")
    print("1. Smart Light")
    print("2. Temperature Sensor")
    print("3. Security Camera")
    print("4. Exit")

    choice = input("Choose a device (1-4): ")
    if choice == "1":
        while True:
            print("\n--- Smart Light Menu ---")
            print("1. Turn ON")
            print("2. Turn OFF")
            print("3. Adjust Brightness")
            print("4. Display Information")
            print("5. Back to Main Menu")

            opt = input("Choose: ")
            if opt == "1":
                light.turn_on()
            elif opt == "2":
                light.turn_off()
            elif opt == "3":
                light.adjust_brightness()
            elif opt == "4":
                light.show_light()
            elif opt == "5":
                break
            else:
                print("Invalid option.")

    elif choice == "2":
        while True:
            print("\n--- Temperature Sensor Menu ---")
            print("1. Turn ON")
            print("2. Turn OFF")
            print("3. Set Temperature")
            print("4. Display Information")
            print("5. Back to Main Menu")

            opt = input("Choose: ")
            if opt == "1":
                sensor.turn_on()
            elif opt == "2":
                sensor.turn_off()
            elif opt == "3":
                sensor.set_temperature()
            elif opt == "4":
                sensor.show_temperature()
            elif opt == "5":
                break
            else:
                print("Invalid option.")

    elif choice == "3":
        while True:
            print("\n--- Security Camera Menu ---")
            print("1. Turn ON")
            print("2. Turn OFF")
            print("3. Start Recording")
            print("4. Stop Recording")
            print("5. Display Information")
            print("6. Back to Main Menu")

            opt = input("Choose: ")
            if opt == "1":
                camera.turn_on()
            elif opt == "2":
                camera.turn_off()
            elif opt == "3":
                camera.start_recording()
            elif opt == "4":
                camera.stop_recording()
            elif opt == "5":
                camera.show_camera()
            elif opt == "6":
                break
            else:
                print("Invalid option.")

    elif choice == "4":
        print("\nThank you for using the Smart Device Management System. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter 1, 2, 3 or 4.")
```
